#### Excercise1 : program to get the difference between a given number and 17. if the number is greater than 17 return double the absolute diffrence


```python
x=float(input("type in any number:"))
y=x-17
if x>17:
    z=y*2
    print(z)
else:
    x<17
    y=x-17
    print(y)
```

#### Excercise2: Python program to calculate the sum of three given numbers . if the values are equal then return thrice of their sum


```python
x=float(input("type in any number:"))
y=float(input("type in any number:"))
z=float(input("type in any number:"))
if x!=y!=z:
    f=x+y+z
    print(f)
else:
    x=y=z
    f=x+y+z
    s=f*3
    print(s)
```

#### Excercise3: python program which will return true if the two given interger values are equall or their sum or diffrence is 5


```python

x=int(input("type in a number"))
y=int(input("type in a number"))
if x==y or x+y==5 or x-y==5:
    print("TRUE")
else:
    print("FALSE")
```

#### Excersice 4: python program to sort three intergers withouth using  conditional statement and loops


```python

import statistics as np
x=int(input("type in a number"))
y=int(input("type in a number"))
z=int(input("type in a number"))
numbers=(x,y,z)
a=min(numbers)
b=np.median(numbers)
c=max(numbers)
print(a,b,c)
```

#### Exercise 5: A pyhton function that takes a positive interger and returns the sum of the cube of all the positive intergers smaller than the specified number
