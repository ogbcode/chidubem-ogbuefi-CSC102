```python
#grid usage
import numpy as np
import matplotlib.pyplot as plt
# creating an empty list
x_values = []
y_values=[]
  
# number of elemetns as input
x_y_no = int(input("Enter number of elements in x and y : "))
titlee=input("what will be the title of your graph:")
ylab=input("what will be the title of your y axis:")
xlab=input("what will be the title of your x axis:")
# iterating till the range
for i in range(0,x_y_no):
    ele = int(input("x value =:"))
    x_values.append(ele)
    yelement=int(input("y value=:"))
    y_values.append(yelement)
print("your x values are",x_values)
print("your y values are",y_values)
#for labelling the side
x = np.array(x_values)
y = np.array(y_values)

plt.scatter(x, y,marker="o")
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'blue','size':20}

plt.title(titlee, fontdict = font1)
plt.xlabel(xlab, fontdict = font2)
plt.ylabel(ylab, fontdict = font2)
plt.grid(color = 'green', linestyle = '--', linewidth = 1)
plt.show()

```
